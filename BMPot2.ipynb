```python
import matplotlib.pyplot as plt
import numpy as np
import math


U=[]
r=[]
e=1.602*(10**(-19)) 
ep=8.84*(10**(-12))
a=3.33*(10**(10))
b=3.237*(10**(-16))
k=0
for i in range (160, 451,10):
    r_pm=i*(10**(-12))
    r.append(i)
    u=(-(e*e)/(4*math.pi*ep*r_pm))+(b*math.exp(-(a*r_pm)))
    print(u)
    U.append(u)
fig1=plt.figure()
axes = plt.gca()
axes.set_ylim([(-1*(10**(-18))),(0.2*(10**(-18)))])
axes.set_xlim([0,500])
plt.grid(color='c', linestyle='--', linewidth=0.8)
plt.ylabel("U (in kJ/mol)")
plt.xlabel("r (in pm)")
plt.plot(r,U,c='b')
plt.plot(r,U,'mo')
plt.savefig("U_r.png",format="png")
plt.show()
```

```python
MyFile=open('output.txt','w')
MyFile.write("r(in pm)"+"  "+"U "
k=0
for element in U:
    MyFile.write(str(r[k])+"      "+str(element))
    MyFile.write('\n')
    k=k+1
MyFile.close()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math
r=[]
REP=[]
ATT=[]
e=1.602*(10**(-19))
ep=8.84*(10**(-12))
a=3.33*(10**(10))
b=3.237*(10**(-16))
k=0
for i in range (180, 280,2):
    r_pm=i*(10**(-12))
    r.append(i)
    att=((e*e)/(4*math.pi*ep*r_pm*r_pm))
    rep=(a*b*math.exp(-(a*r_pm)))
    ATT.append(att)
    REP.append(rep)
fig2=plt.figure()
plt.grid(color='c', linestyle='--', linewidth=0.8)
axes = plt.gca()
axes.set_ylim([0,(3*(10**(-8)))])
axes.set_xlim([0,300])
plt.ylabel("e^2/(4πƐr^2)  abexp(-ar)")
plt.xlabel("r (in pm)")
plt.plot(r,ATT,c='b',label="e^2/(4πƐr^2)")
plt.plot(r,ATT,'bo')
plt.plot(r,REP,c='r',label="abexp(-ar)")
plt.plot(r,REP,'ro')
plt.legend()
plt.savefig("ATRE.png",format="png")
plt.show()
```

```python
MyFile=open('output.txt','w')
k=0
for element in ATT:
    MyFile.write(str(r[k])+"    "+str(element)+"    "+str(REP[k]))
    MyFile.write('\n')
    k=k+1
MyFile.close()
```
